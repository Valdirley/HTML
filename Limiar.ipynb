LImiarização

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/Milho1.jpeg')
plt.imshow(img);
```

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb);
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY);
plt.imshow(gray, cmap = 'gray');
```

```python
limiar_min = 84.8 #@param {type: "slider", min:0, max:255, step:0.1}
limiar_max = 255 #@param {type: "slider", min:0, max:255, step:0.1}
val, busca_limiar = cv2.threshold(gray, limiar_min, limiar_max, cv2.THRESH_BINARY)
plt.imshow(busca_limiar, cmap='gray');
```
